```python
import pandas as pd
```

```python
df = pd.read_csv('/content/hasil_saya (1).csv')
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
```

```python
text = df['content']
```

```python
text
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(text)
print(tfidf_matrix)
```

# Interpretasi
Format "(i, j) x.y" digunakan untuk menunjukkan skor TF-IDF suatu istilah dalam suatu dokumen. Mari kita jelaskan interpretasi yang lebih terperinci:

Contoh "(0, 3251) 0.1055299758616986": (pada baris 1)

"(0, 3251)" mengindikasikan bahwa ini adalah skor TF-IDF untuk istilah yang terdapat di dokumen 0 dan memiliki indeks 3251 dalam representasi fitur.
"0.1055299758616986" adalah nilai skor TF-IDF untuk istilah tersebut dalam dokumen 0. Semakin tinggi nilai ini, semakin penting atau khas istilah tersebut dalam dokumen tersebut.
Dengan kata lain, output ini memberikan informasi tentang seberapa penting atau berbedanya suatu istilah dalam setiap dokumen. Semakin tinggi skor TF-IDF suatu istilah dalam suatu dokumen, semakin khas atau penting istilah tersebut dalam dokumen tersebut. Ini membantu dalam proses pengambilan informasi, mengidentifikasi istilah-istilah yang memiliki dampak signifikan dalam setiap dokumen.







```python
nama_fitur=tfidf.get_feature_names_out()
nama_fitur_baru=pd.DataFrame(nama_fitur)
nama_fitur_baru
```

# Interpretasi
Angka-angka tiap kolom yang kebanyakan bernilai 0 adalah akibat tidak adanya representasi kata tersebut pada kalimat yang ditampilkan. Kolom yang dihasilkan juga terlalu banyak karena pada kasus ini data belum dibersihkan sehingga banyak noise seperti '00','000', '0sebelumnya', dll.
Seharusnya jika kita dapat melihat semua fitur/kolom yang ada maka kata-kata yang muncul pada kalimat pertama akan memiliki nilai pada kolom kata tersebut, nilainya didasarkan dengan rumus TF-IDF yaitu:
**TF-IDF(t, d) = TF(t, d) * IDF(t)**,
dengan

**TF(t, d) = (jumlah kemunculan kata t dalam dokumen d) / (jumlah kata total dalam dokumen d)**,
dan
**IDF(t) = log(N / (jumlah dokumen yang mengandung kata t))**
