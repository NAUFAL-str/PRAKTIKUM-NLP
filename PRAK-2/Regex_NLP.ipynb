```python
from faker import Faker

fake = Faker()

jumlah_data = 10

dummy_data = []

for _ in range(jumlah_data):
    nama = fake.name()
    nomor_telepon = fake.phone_number()
    alamat = fake.address()
    email = fake.email()

    data = f"Nama: {nama}\nNomor Telepon: {nomor_telepon}\nAlamat: {alamat}\nEmail: {email}\n"
    dummy_data.append(data)

file_name = "dummy_data.txt"
with open(file_name, "w") as file:
    for data in dummy_data:
        file.write(data + "\n")

print(f"{jumlah_data} data dummy telah disimpan dalam {file_name}.")
```

```python

file_path = "dummy_data.txt"

with open(file_path, "r") as file:

    text = file.read()

print(text)
```

```python

import re

def extract_info(text):
    # Regex pattern
    name_pattern = r'Nama:\s(.*?)\n'
    phone_pattern = r'Nomor Telepon:\s(.*?)\n'
    address_pattern = r'Alamat:\s(.*?)\n'
    email_pattern = r'Email:\s(.*?)\n'
    email_name_pattern = r'Email:\s(.*?)@'
    #re.findall
    names = re.findall(name_pattern, text)
    phones = re.findall(phone_pattern, text)
    addresses = re.findall(address_pattern, text)
    emails = re.findall(email_pattern, text)
    email_names = re.findall(email_name_pattern, text)
    # Menghilangkan white space,- ,dll serta membuatnya dalam dictionary
    info_list = []
    for i in range(len(names)):
        info_dict = {
            "Name": names[i].strip(),
            "Phone": re.sub(r'\D', '', phones[i]).strip(),
            "Address": addresses[i].strip(),
            "Email": emails[i].strip(),
            "EmailName": email_names[i].strip(),
        }
        info_list.append(info_dict)

    return info_list


info_list = extract_info(text)

for info in info_list:
    print("Name:", info["Name"])
    print("Phone:", info["Phone"])
    print("Address:", info["Address"])
    print("Email:", info["Email"])
    print("Email Name:", info["EmailName"])
    print()
```

```python
import pandas as pd

df =pd.DataFrame(info_list)
df
```
